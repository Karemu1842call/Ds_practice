GDP Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
country_codes = pd.read_csv('country_codes.csv')
gdpdata = pd.read_csv('gdp_data.csv')
merged_table =pd.merge(country_codes,gdpdata,on='country_code', how='inner')
merged_table
```

A Graph of Worlds GDP Growth Since 1960

```python
# Group by "years" and calculate the total "value" for each year
yearly_total_gdp = gdpdata.groupby("year")["value"].sum().reset_index()

# Sort the DataFrame by total "value" in descending order
yearly_total_gdp = yearly_total_gdp.sort_values(by="value", ascending=False)

# Create a bar plot
plt.figure(figsize=(15, 7))
sns.barplot(x="year", y="value", data=gdpdata, palette="viridis")
plt.xlabel("Year")
plt.ylabel("Total GDP}")
plt.title("World GDP Groth Over the Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
russia_data = merged_table.loc[merged_table['country_code']=='RUS']
plt.figure(figsize=(15, 7))
sns.barplot(x="year", y="value", data=russia_data, palette="viridis")
plt.xlabel("Year")
plt.ylabel("Total GDP}")
plt.title("Russia GDP Groth Over the Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
us_gdp = merged_table.loc[merged_table['country_code']=='USA']
plt.figure(figsize=(15, 7))
sns.barplot(x="year", y="value", data=us_gdp, palette="viridis")
plt.xlabel("Year")
plt.ylabel("Total GDP}")
plt.title("Russia GDP Groth Over the Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Kenyas Table

```python
kenyadata= gdpdata.loc[gdpdata['country_code']== 'KEN']
kenya_gdp = kenyadata.groupby("year")["value"].sum().reset_index()

# Sort the DataFrame by total "value" in descending order
kenya_gdp = kenya_gdp.sort_values(by="value", ascending=False)
kenya_gdp
```

```python
plt.figure(figsize=(15, 7))
sns.barplot(x="year", y="value", data=kenya_gdp, palette="viridis")
plt.xlabel("Year")
plt.ylabel("Total GDP}")
plt.title("Kenya GDP Groth Over the Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
years = sorted(merged_table['year'].unique().tolist())
def make_plot_data() -> dict:
    plot_df = merged_table
    data = [dict(type='choropleth', locations = plot_df[plot_df['year'] == year]['country_code'], 
                 z=plot_df[plot_df['year'] == year]['value'],
                 hovertext=plot_df[plot_df['year'] == year]['country_name']) for year in years]
    steps = [dict(method='restyle', args=['visible', [other == year for other in years]], label=year) for year in years]
    layout = dict(geo=dict(scope='world'), sliders=[dict(active=0, pad={'t': 1}, steps=steps)], title="World GDP by Country Over the Years")
    return dict(data=data, layout=layout)

from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)
iplot(make_plot_data())
```
