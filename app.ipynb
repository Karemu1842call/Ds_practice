GDP Data

```python
import pandas as pd
import matplotlib.pyplot as plt
country_codes = pd.read_csv('country_codes.csv')
gdpdata = pd.read_csv('gdp_data.csv')
gdpdata
```

A Graph of Kenyas GDP Since 1960

```python
plt.figure(figsize=(12, 6))
plt.plot(gdpdata['year'], gdpdata['value'] )
plt.title('GDP Over Time in Kenya')
plt.xlabel('Year')
plt.ylabel('GDP Value')
plt.grid(True)

# Show the plot
plt.show()
```

Average GDP based on years.

```python
country_data= gdpdata.groupby('year')
average_years = country_data['value'].mean()
average_years
```
